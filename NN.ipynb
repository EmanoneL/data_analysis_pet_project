```python
import pandas
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets
```

```python
display(X.head(10), y.head(10))
```

```python
encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)
dummy_y = to_categorical(encoded_Y)
```

```python
def print_nn_plots(history):
  plt.subplot(211)
  plt.plot(history.history['accuracy'])
  plt.plot(history.history['val_accuracy'])
  plt.title('Model accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Test'], loc='upper left')
  plt.show()

  plt.subplot(212)
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Test'], loc='upper left')
  plt.show()
```

```python
def fit_model(model, optim='adam',los='categorical_crossentropy', metric=['accuracy'], epoch=75, batch_siz=10, validation_spli=0.1):
  model.compile(optimizer=optim,loss=los, metrics=metric)
  history = model.fit(X, dummy_y, epochs=epoch, batch_size=batch_siz, validation_split=validation_spli, verbose=True)
  train_loss, train_acc = model.evaluate(X, dummy_y)
  print(f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_acc:.4f}")

  print_nn_plots(history)
```

```python
neirons = (4, 16, 32, 68)
```

```python
model = Sequential()

for i in neirons:
  model.add(Dense(i, activation='relu'))
  model.add(Dense(3, activation='softmax'))
  print(f'Количество слоев: 2. Нейронов на 1 слое:{i}')
  fit_model(model)
```

```python
model = Sequential()

for i in neirons:
  for j in neirons:
    model.add(Dense(i, activation='relu'))
    model.add(Dense(j, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    print(f'Количество слоев: 3. Нейронов на 1 слое:{i} Нейронов на 2 слое:{j}')
    fit_model(model)
```

```python
# Перебраем параметры обучения

model = Sequential()
model.add(Dense(16, activation='relu'))
model.add(Dense(3, activation='softmax'))

batch_sizes = [8, 16, 32]
epochs = [50, 150, 250]
vals = [0.15, 0.2, 0.3]
print(batch_sizes, epochs, vals)

for val in vals:
  for batch in batch_sizes:
    for epoch in epochs:
      print(f'Модель с параметрами epoch={epoch}, batch_size={batch}, validation_split={val}')
      fit_model(model, batch_siz=batch, epoch=epoch, validation_spli=val)
```
